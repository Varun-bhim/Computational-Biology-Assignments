```r
install.packages("BiocManager")
BiocManager::install("AnnotationHub")
BiocManager::install("ensembldb")
```

```r
library(AnnotationHub)
library(ensembldb)
```

```r
ah <- AnnotationHub()
edb <- ah[["AH119325"]]
```

```r
tx <- transcripts(edb, filter = ~ gene_name == "OPN1LW")
tx

write.table(tx,
            file = "OPN1LW_transcripts.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```

```r
pr <- proteins(edb, filter = ~ gene_name == "OPN1LW")
pr

write.table(pr,
            file = "OPN1LW_proteins.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```

```r
tx_ids <- pr$tx_id
tx_ids

tx_gr <- transcripts(edb, filter = ~ tx_id %in% tx_ids)
tx_gr

mapping_df <- as.data.frame(tx_gr)

write.table(mapping_df,
            file = "OPN1LW_protein_genome_map.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```
